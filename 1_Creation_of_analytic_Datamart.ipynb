I don't know why when you use a large amount of data the function pd.to_datatime the time to do the conversion is too long (approx 5 minutes). To solve this minor problem, we can use the next function that makes the process in less than 15 sec.

```python
#df['event_time'] = pd.to_datetime(df['event_time'])
#This method consum a lot of time (aprox 5 minutes) compared with the function datetime_quickly
```

```python
def datetime_quickly(df, formato):
    '''
    This function help us to convert very quickly a big numbers of registers in a datetime format
    
    '''
    def split_date(date):
        division = date.split()
        date = division[0]
        time = division[1]
        cadena = date + ' ' + time
        return (cadena)
    result = pd.to_datetime(df.apply( lambda x: split_date(x)), format = formato)
    return(result)
        
```

```python
formato = '%Y-%m-%d %H:%M:%S'
df.event_time = datetime_quickly(df['event_time'], formato)
df.info()
```

## Data transformation

We are going to create 3 types of new variables

* Extract components
* Calendar variables: Local holidays (Russia)
* Exogenous indicators: Days not necessarily holidays but with commercial interest: Black Friday, Cyber Monday, Epiphany, Valentine's Day.

### Components of the date

```python
def date_components(dataframe):
    ''' 
    This function split the components of an index date in their parts.
    '''
    date = dataframe.index.date
    year = dataframe.index.year
    month = dataframe.index.month
    day = dataframe.index.day
    hour = dataframe.index.hour
    minute = dataframe.index.minute
    second = dataframe.index.second
    
    
    return(pd.DataFrame({'date':date, 'year':year,'month':month, 'day':day, 'hour':hour, 'minute':minute, 'second':second}))
```

```python
df = pd.concat([df.reset_index(), date_components(df)], axis = 1)
```

### Calendar vairables: holidays

```python
import holidays
```

```python
holidays_RU = holidays.RU(years = 2020)
holidays_RU
```

```python
df['holidays'] = df.date.apply(lambda x: 1 if (x in holidays_RU) else 0)
```

```python
df[df.holidays == 1].date.value_counts().sort_index()
```
